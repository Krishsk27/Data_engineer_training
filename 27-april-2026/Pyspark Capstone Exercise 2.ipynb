```python
from pyspark.sql.functions import lit, concat, when
```

```python
p_mod = patients.withColumn("hospital_name", lit("BotCampus Hospital")) \
                .withColumn("country", lit("India")) \
                .withColumn("patient_label", concat(col("name"), lit(" - "), col("city"))) \
                .withColumn("senior_patient_flag", col("age") > 40)
```

```python
b_mod = bills.withColumn("bill_with_tax", col("bill_amount") * 1.18) \
             .withColumn("bill_in_thousands", col("bill_amount") / 1000) \
             .withColumn("high_bill_flag", col("bill_amount") > 1500)
```

```python
patients.withColumn("age_category",
    when(col("age") < 30, "Young").when(col("age") < 50, "Adult").otherwise("Senior")).show()
```

```python

bills.withColumn("bill_category",
    when(col("bill_amount") > 1500, "High").when(col("bill_amount") >= 1000, "Medium").otherwise("Low")).show()
```

```python
appointments.withColumn("priority",
    when(col("status") == "Pending", "High").when(col("status") == "Completed", "Normal").otherwise("Low")).show()
```

```python
from pyspark.sql.window import Window
from pyspark.sql.functions import rank, row_number, dense_rank
```

```python
city_window = Window.partitionBy("city").orderBy(col("age").desc())
spec_window = Window.partitionBy("specialization").orderBy(col("consultation_fee").desc())
```

```python
patients.withColumn("rank", rank().over(city_window)).show()
```

```python
patients.withColumn("rn", row_number().over(city_window)).filter(col("rn") == 1).show()
```

```python
doctors.withColumn("fee_rank", dense_rank().over(spec_window)).show()
```

```python
pay_window = Window.partitionBy("payment_mode").orderBy("bill_id")
bills.withColumn("running_total", sum("bill_amount").over(pay_window)).show()
```
