```python
import ipywidgets as widgets
from IPython.display import display
```

```python
name_widget = widgets.Text(
value="Sureshkrishna",
description="Name:"

)

display(name_widget)
```

```python
dept_widget = widgets.Dropdown(
options=["IT", "HR", "Finance"],
value="IT",
description="Dept:")

display(dept_widget)
```

```python
salary_slider = widgets.IntSlider(
value=60000,
min=30000,
max=100000,
step=5000,
description="Salary:"

)

display(salary_slider)
```

```python
button = widgets.Button(description="Show Values")

def on_button_click (b):
  print("Name:", name_widget.value)
  print("Department:", dept_widget.value)
  print("Salary:", salary_slider.value)

button.on_click(on_button_click)
display(button)
```
